```python
import pandas as pd 
import matplotlib.pyplot as mpl
df = pd.read_csv('vgsales.csv', sep=',')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.nunique()
```

```python
df=df.dropna()
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['Year']=pd.to_datetime(df['Year'])
df.info()
```

```python
df['Genre'].unique()
```

```python
df['Genre'].value_counts()
```

```python
import seaborn  as sns
df.head()
```

```python
sns.countplot(x='Genre',data=df,order=df['Genre'].value_counts().index)
mpl.xticks(rotation=90); 
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum().reset_index().sort_values(by = 'Global_Sales', ascending = False)
genre_sales['percentage_of_total_sales'] = round(genre_sales['Global_Sales'] / genre_sales['Global_Sales'].sum() * 100, 1)
genre_sales
```

```python
mpl.figure(figsize = (15, 10))

fig, ax = mpl.subplots()
ax.pie(genre_sales['Global_Sales'], labels=genre_sales['Genre'], autopct='%1.1f%%')
ax.axis("equal")
mpl.title('Sales percentage by genre', x= 0.5 , y= 1.2 )
mpl.show()
```

```python
top = df.groupby('Name')['Global_Sales'].sum().sort_values(ascending = False).reset_index().head(20)
top
```

```python
mpl.figure(figsize = (12, 6))
sns.barplot(data = top, x = 'Name', y = 'Global_Sales').set(title = 'Top of popular games')
mpl.ylabel('Global sales')
mpl.xlabel('Name')
mpl.xticks(rotation = 90)
mpl.show()
```

```python
df['Year'] = pd.to_numeric(df['Year'])
year_games = df['Year'].value_counts().reset_index().sort_values(by = 'count', ascending = False)
year_games
```

```python
mpl.figure(figsize = (12, 6))
sns.barplot(data = year_games, x = 'Year', y = 'count').set(title = 'Dynamics of the count of games by year')
mpl.xticks(rotation = 90)
mpl.show()
```

```python
game_on_platform = df['Platform'].value_counts().reset_index()
game_on_platform
```

```python
mpl.figure(figsize = (12, 6))
sns.barplot(data = game_on_platform, x = 'Platform', y = 'count').set(title = 'Count of games released on different platforms')
mpl.ylabel('Count of games released')
mpl.xlabel('Platform name')
mpl.xticks(rotation = 90)
mpl.show()
```

```python
ranking_platforms = df.groupby('Platform')['Global_Sales'].sum().reset_index().sort_values(by = 'Global_Sales', ascending = False)
ranking_platforms
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()


mpl.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
mpl.title('Correlation of Video Game Sales')
mpl.show()
```
